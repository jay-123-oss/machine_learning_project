```python
from sklearn.preprocessing import MinMaxScaler

# Keep only Close price
# Convert to numpy array and reshape to 2D
dataset = df_main.values.reshape(-1, 1)

# Normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)

print(scaled_data[:5])
```

```python
import numpy as np

# Define how many days of past data the model should look at
time_step = 60

X = []
y = []

# Loop through the data to create shifting sequences
for i in range(time_step, len(scaled_data)):
    X.append(scaled_data[i - time_step:i, 0])  # Past 'time_step' days
    y.append(scaled_data[i, 0])                 # The next day's price (target)

# Convert lists to numpy arrays
X, y = np.array(X), np.array(y)

# Reshape X to be [samples, time_steps, features] required for LSTM
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

print(f"X shape (Samples, Time Steps, Features): {X.shape}")
print(f"y shape (Samples,): {y.shape}")
```

```python
X[2
]
```

```python
# Calculate the split point (80% for training)
train_size = int(len(X) * 0.80)

# Chronological split
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

print("=== Split Results ===")
print(f"X_train shape: {X_train.shape} | y_train shape: {y_train.shape}")
print(f"X_test shape:  {X_test.shape}  | y_test shape:  {y_test.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

# Initialize the Sequential model
model = Sequential()

# 1st LSTM Layer (Must have return_sequences=True since another LSTM layer follows)
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))  # Prevents overfitting

# 2nd LSTM Layer (return_sequences=False because the next layer is a Dense layer)
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

# Output Layer (1 neuron for predicting the next day's price)
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Summary of the architecture
model.summary()
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import os

# 1. Checkpoint ka path
checkpoint_path = "/kaggle/working/lstm_model_checkpoint.keras"

# 2. Checkpoint Callback (Best weights save karne ke liye)
checkpoint = ModelCheckpoint(
    filepath=checkpoint_path,
    monitor='val_loss',
    save_best_only=True,
    mode='min',
    verbose=1
)

# 3. Early Stopping Callback (Training rokne ke liye)
early_stop = EarlyStopping(
    monitor='val_loss',   # Validation loss ko check karega
    patience=5,           # Agar 5 epochs tak loss nahi sudhra, toh training stop kar dega
    restore_best_weights=True, # Training rukne par sabse best wali epoch ke weights load kar dega
    mode='min',
    verbose=1
)

# Model ko train karein (Dono callbacks ko list mein pass karein)
history = model.fit(
    X_train, y_train,
    epochs=50,            # Ab aap bina dar ke epochs badha sakte hain
    batch_size=32,
    validation_data=(X_test, y_test),
    callbacks=[checkpoint, early_stop], # Dono shields active hain
    verbose=1
)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# 1. Test data par predictions nikalein
predictions = model.predict(X_test)

# 2. Data ko wapas original price format mein Inverse Transform karein
# Kyunki scaler 2D output expect karta hai, hum predictions aur y_test dono ko transform karenge
predictions_unscaled = scaler.inverse_transform(predictions)
y_test_unscaled = scaler.inverse_transform(y_test.reshape(-1, 1))

# 3. Accuracy Metrics Calculate Karein
mse = mean_squared_error(y_test_unscaled, predictions_unscaled)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_unscaled, predictions_unscaled)

print("============ MODEL ACCURACY ============")
print(f"Mean Absolute Error (MAE): ${mae:.2f}")
print(f"Root Mean Squared Error (RMSE): ${rmse:.2f}")
print("========================================")

# 4. Actual vs Predicted Prices ka Graph Plot Karein
plt.figure(figsize=(14, 5))
plt.plot(y_test_unscaled, color='blue', label='Actual Apple Stock Price')
plt.plot(predictions_unscaled, color='red', linestyle='--', label='Predicted Apple Stock Price')
plt.title('Apple Stock Price Prediction - Actual vs Predicted')
plt.xlabel('Time (Days)')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# 1. Pichle 60 din ka scaled data uthaein (Base for future prediction)
# Hum data ko sahi se list mein convert kar rahe hain taaki naye points append kar sakein
current_batch = list(scaled_data[1000:1060, 0])
future_predictions_scaled = []

# 2. Loop chalayein agle 10 din ke liye
for i in range(60):
    # Current batch ko array banakar LSTM ke shape (1, 60, 1) mein dalein
    X_input = np.array(current_batch[-60:]).reshape(1, 60, 1)
    
    # Agle din ka price predict karein
    next_day_pred = model.predict(X_input, verbose=0)
    
    # Prediction ko safe karein (scaled format mein)
    future_predictions_scaled.append(next_day_pred[0, 0])
    
    # Is nayi prediction ko batch mein jod dein taaki agla din predict ho sake
    current_batch.append(next_day_pred[0, 0])

# 3. Scaled predictions ko wapas original dollar ($) price mein convert karein
future_predictions = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1, 1))

# 4. Results ko saaf-saaf print karein
print("\n================== NEXT 10 DAYS PREDICTIONS ==================")
output = []
for day, price in enumerate(future_predictions, 1):
    print(f"Day {day}: ${price[0]:.2f}")
    output.append(price)
print("==============================================================")
```
